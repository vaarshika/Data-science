```python
print('the percentange of missing values in LB',cardio['LB'].isnull().sum() / (len(cardio))* 100)
print('the percentange of missing values in AC',cardio['AC'].isnull().sum() / (len(cardio))* 100)
print('the percentange of missing values in DS',cardio['DS'].isnull().sum() / (len(cardio))* 100)
print('the percentange of missing values in DP',cardio['DP'].isnull().sum() / (len(cardio))* 100)
print('the percentange of missing values in MLTV',cardio['MLTV'].isnull().sum() / (len(cardio))* 100)
print('the percentange of missing values in Width',cardio['Width'].isnull().sum() / (len(cardio))* 100)
print('the percentange of missing values in Tendency',cardio['Tendency'].isnull().sum() / (len(cardio))* 100)
print('the percentange of missing values in NSP',cardio['NSP'].isnull().sum() / (len(cardio))* 100)
```

```python
LB_mean = cardio['LB'].mean()
print('the average of LB in cardio : ', LB_mean)
AC_mean = cardio['AC'].mean()
print('the average of AC in cardio : ', AC_mean)
DS_mean = cardio['DS'].mean()
print('the average of DS in cardio : ', DS_mean)
DP_mean = cardio['DP'].mean()
print('the average of DP in cardio : ', DP_mean)
MLTV_mean = cardio['MLTV'].mean()
print('the average of MLTV in cardio : ', MLTV_mean)
Width_mean = cardio['Width'].mean()
print('the average of Width in cardio : ', Width_mean)
Tendency_mean = cardio['Tendency'].mean()
print('the average of Tendency in cardio : ', Tendency_mean)
NSP_mean = cardio['NSP'].mean()
print('the average of NSP in cardio : ', NSP_mean)
```

```python
cardio['LB'] = cardio['LB'].fillna(LB_mean)
cardio['AC'] = cardio['AC'].fillna(AC_mean)
cardio['DS'] = cardio['DS'].fillna(DS_mean)
cardio['DP'] = cardio['DP'].fillna(DP_mean)
cardio['MLTV'] = cardio['MLTV'].fillna(MLTV_mean)
cardio['Width'] = cardio['Width'].fillna(Width_mean)
cardio['Tendency'] = cardio['Tendency'].fillna(Tendency_mean)
cardio['NSP'] = cardio['NSP'].fillna(NSP_mean)
```

```python
import warnings
warnings.filterwarnings('ignore')
from pandas.plotting import scatter_matrix
scatter_matrix(cardio , figsize = (15,30) , diagonal = 'hist' , color = 'green')
```

```python
def remove_outliers(data):
  #numerical_columns = data.select_dtypes(include =['int64','float64']).columns
  for i in data:
    q1 = data[i].quantile(0.25)
    q3 = data[i].quantile(0.75)
    iqr = q3 -q1
    ul = q3 + (1.5 * iqr)
    ll = q1 - (1.5 * iqr)
    data = data[(data[i]<=ul) & (data[i]>=ll)]
  return data

r_out_data = remove_outliers(cardio)
r_out_data
```

```python
plt.figure(figsize = (10,6))
#correlation matrix
correlation_matrix = r_out_data.corr()
sns.heatmap(correlation_matrix , annot = True , cmap = 'coolwarm',fmt = '.2f')
```
