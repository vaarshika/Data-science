```python
def remove_outliers(data):
  numerical_columns = data.select_dtypes(include = ['int64','float64'])
  for i in numerical_columns:
    Q1 = data[i].quantile(0.25)
    Q3 = data[i].quantile(0.75)
    iqr = Q3 -Q1
    ul = Q3 + (1.5 * iqr)
    ll = Q1 - (1.5 * iqr)

    data = data[(data[i] >= ll) & (data[i] <= ul)]
    return data

remove_out = remove_outliers(wine)
remove_out
```

# Task 2: Dimensionality Reduction with PCA:

```python
from sklearn.preprocessing import StandardScaler
norm = StandardScaler()
scaled_data = norm.fit_transform(wine)
scaled_data
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA()
pca_values = pca.fit_transform(scaled_data)
pca_values
```

```python
pca_comp = pd.DataFrame(pca_values)
pca_comp
```

```python
var = pca.explained_variance_ratio_
var
```

```python
plt.plot(var,color = "green")
```

```python
#cumulative variance
import numpy as np
var1 = np.cumsum(np.round(var,decimals = 4) * 100)
var1
```

```python
pca = PCA(n_components = 3)
pca_values = pca.fit_transform(scaled_data)
pca_values
```

```python
var = pca.explained_variance_ratio_
var
```

```python
pca.components_
```

```python
x = pca_values[:,0]
y = pca_values[:,1]
plt.scatter(x,y)
```

```python
final_output = pd.concat([pd.DataFrame(pca_values[:,0:],columns = ['pc1','pc2','pc3']),wine[['Type']]],axis =1)
final_output
```

```python
import seaborn as sns
sns.scatterplot(data = final_output , x = 'pc1' ,y = 'pc2' , hue = 'Type' , style = 'Type', size = "Type")
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
wcss = []
for i in range(1,10):
  kmeans = KMeans(n_clusters = i , random_state = 0)
  kmeans.fit(scaled_data)
  wcss.append(kmeans.inertia_)
print(wcss)
plt.plot(range(1,10),wcss)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 3,random_state = 0)
kmc = kmeans.fit(scaled_data)
kmc
```

```python
y_kmc = kmc.fit_predict(scaled_data)
y_kmc
```
